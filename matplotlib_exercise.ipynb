```python
df_1 = pd.read_csv("data/hist_data.csv")
plt.hist(df_1["aデータ"], bins=30, color="orange", alpha=0.5, label="aデータ")
plt.hist(df_1["bデータ"], bins=30, color="cyan", alpha=0.5, label="bデータ")
plt.legend()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_2 = pd.read_csv("data/scatter_data.csv")
class_0 = df_2[df_2["class"]==0]
class_1 = df_2[df_2["class"]==1]
class_2 = df_2[df_2["class"]==2]
plt.scatter(class_0["length (cm)"], class_0["width (cm)"], color="red", label="class0")
plt.scatter(class_1["length (cm)"], class_1["width (cm)"], color="blue", label="class1")
plt.scatter(class_2["length (cm)"], class_2["width (cm)"], color="green", label="class2")
plt.title("クラス0～2のlengthとwidthに関する散布図")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.legend()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_3 = pd.read_csv("data/line_data.csv")
period = df_3["期間"]
count = df_3["件数"]
cost = df_3["費用"]

plt.figure()
plt.plot(period, count, marker="o")
plt.grid(True)
plt.xticks(rotation=30)
plt.title("期間ごとの件数の推移")
plt.xlabel("期間")
plt.ylabel("件数")

plt.figure()
plt.bar(period, cost)
plt.grid(True)
plt.xticks(rotation=30)
plt.title("期間ごとの費用の推移")
plt.xlabel("期間")
plt.ylabel("費用(円)")
```
